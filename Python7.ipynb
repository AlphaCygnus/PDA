```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("stock_data.csv")
```

```python
display(data.head(10))
```

```python
plt.scatter(data['Open'], data['Close'])
plt.title("Scatter plot")
plt.xlabel('Open')
plt.ylabel('Close')
plt.show()
```

```python
plt.scatter(data['Open'], data['Low'], c=data['Low'], s=data['Close'])
plt.title("Scatter Plot")
plt.xlabel('Open')
plt.ylabel('Low')
plt.colorbar()
plt.show()
```

```python
plt.plot(data['Open'])
plt.plot(data['Low'])
plt.title("Scatter Plot")
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

```python
plt.bar(data['Date'], data['Open'])
plt.title("Bar Chart")
plt.xlabel('Date')
plt.ylabel('Open')
plt.show()
```

```python
plt.hist(data['Close'])
plt.title("Histogram for the data")
plt.show()
```

```python
import seaborn as sns
sns.lineplot(x="Open", y="Close", data=data)
plt.title('Title using Matplotlib Function')
plt.show()
```

```python
sns.histplot(x='Open', data=data, kde=True, hue='Close')
plt.show()
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook
```
