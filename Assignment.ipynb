```python
import pandas as pd
```

```python
data = {'Name': ['Brendan', 'May', 'Gustav',
                 'Cynthia', 'Rupert', 'Brando', 'Rachel'],
        'Age': [12, 13, 12, 14, 12, 13, 11],
        'Gender': ['M', 'F', 'M', 'F', 'M', 'M', 'F'],
        'Marks': [88, 91, 'NaN', 99, 65, 'NaN', 63]}
 
df = pd.DataFrame(data)
 
df
```

```python
c = avg = 0
for ele in df['Marks']:
    if str(ele).isnumeric():
        c += 1
        avg += ele
        
if c != 0:
    avg = avg/c

df = df.replace(to_replace="NaN",
                value=avg)
 
df
```

```python
df['Gender'] = df['Gender'].map({'M': 0,
                                 'F': 1, }).astype(float)
 
df
```

```python
df = df[df['Marks'] >= 70].copy()
df
```
